## Overview:

Microsoft Excel is a common choice for researchers.  Excel can produce a range of plots - easily - that are useful during the lifetime of the study. However, Excel plots are often not suitable for Journal publication as the resolution of the charts it produces is below the typical **300dpi** requirement of academic publishers. It also makes reviewers grumpy. This tutorial aims to help researchers used to generating plots in Excel transition to using Python's `matplotlib` and `seaborn` modules.
